```python
# splitting dataFrame by Quarter
dfs = dict(tuple(df.groupby('Quarter_ID')))
Quater_ID_1 = dfs[1]
Quater_ID_2 = dfs[2]
Quater_ID_3 = dfs[3]
Quater_ID_4 = dfs[4]
Quater_ID_5 = dfs[5]
```

# Creating a 3d Matrix

```python
def create3DMatrix(quarter):
    Manager_ID = sorted(list(quarter.Manager_ID.unique()))
    Stock_ID =sorted(list(quarter.Stock_ID.unique()))
    dfMatrix = pd.DataFrame(columns = Manager_ID, index=Stock_ID)
    # filling matrix by 0
    dfMatrix = dfMatrix.fillna(0)
    managerid_list = list(quarter.Manager_ID)
    stockid_list = list(quarter.Stock_ID)
    # for every pair of managerid and stock id storing 1 
    for i in range(len(managerid_list)):
        dfMatrix[managerid_list[i]][stockid_list[i]] = dfMatrix[managerid_list[i]][stockid_list[i]]+1
    return dfMatrix
```

```python
newMatrix1 = create3DMatrix(Quater_ID_1)
newMatrix2 = create3DMatrix(Quater_ID_2)
newMatrix3 = create3DMatrix(Quater_ID_3)
newMatrix4 = create3DMatrix(Quater_ID_4)
newMatrix5 = create3DMatrix(Quater_ID_5)
```

# Counting 1's rowWise and splitting if value based on sumOfRow = 1

```python
def rowWise(newMatrix):
    cnt = 0
    for index, row in newMatrix.iterrows():
        count = 0
        temp = (list(row))
        count = temp.count(1)
        if(count >0):
            print(count)
            for i in range(len(temp)):
                temp[i] =  temp[i]/count
        newMatrix.iloc[cnt] = temp
        cnt +=1
    return newMatrix
```

```python
Matrix1 = rowWise(newMatrix1)
Matrix2 = rowWise(newMatrix2)
Matrix3 = rowWise(newMatrix3)
Matrix4 = rowWise(newMatrix4)
Matrix5 = rowWise(newMatrix5)
```
